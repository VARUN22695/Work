##MAP , FILTER, REDUCE

```python
l=[2,3,4,5,6,7,8,9,10,12,13,14]
```

```python
j=filter(lambda x : x%2==0,l)
```

```python
type(j)
```

```python
list(j)
```

```python
d=map(lambda x: x**2,l)
```

```python
list(d)
```

```python
for i in j:
  print(i)
```

```python
k=reduce(lambda x ,y:x*y,l)
```

```python
from urllib.request import urlopen
import matplotlib.pyplot as plt
import numpy as np
import cv2

# Download the image from the URL
response = urlopen("https://apicms.thestar.com.my/uploads/images/2023/07/29/2203255.jpg")
image_array = np.array(bytearray(response.read()), dtype=np.uint8)
img = cv2.imdecode(image_array, -3)

# Display the image
plt.imshow(img)
plt.show()
```
